```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("datos/pinguinos.csv")
df = df.dropna()
df
```

```python
sns.jointplot(
    data=df,
    x="flipper_length_mm",
    y="culmen_length_mm"
)
```

```python
sns.jointplot(
    data=df,
    x="flipper_length_mm",
    y="culmen_length_mm",
    hue="species"
)
```

```python
sns.jointplot(
    data=df,
    x="flipper_length_mm",
    y="culmen_length_mm",
    kind="reg"
)
```
