```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datos/titanic.csv")
df = df.dropna(subset=["Age"], axis="index")
df
```

```python
df.hist(
    column="Age"
)
```

```python
df.hist(
    column="Age",
    bins=20
)
```

```python
df.hist(
    column="Age",
    bins=5
)
```

```python
df.hist(
    column="Age",
    bins=200
)
```

```python
df.hist(
    column="Age",
    by="Sex",
    bins=15
)
```

# KDE

```python
sns.histplot(
    data=df,
    x="Age"
)
```

```python
sns.histplot(
    data=df,
    x="Age",
    kde=True
)
```

```python
sns.histplot(
    data=df,
    x="Age",
    kde=True, 
    bins=5
)
```

```python
sns.histplot(
    data=df,
    x="Age",
    kde=True,
    bins=20
)
```

```python
sns.histplot(
    data=df,
    x="Age",
    kde=True,
    stat="probability"
)
```

```python
sns.histplot(
    data=df,
    x="Age",
    kde=True,
    bins=20,
    stat="count",
    hue="Sex"
)
```

```python
sns.histplot(
    data=df,
    x="Age",
    bins=20,
    cumulative=True,
    fill=True
)
```

```python
sns.histplot(
    data=df,
    x="Age",
    bins=20,
    cumulative=True,
    fill=False,
    element="step"
)
```
