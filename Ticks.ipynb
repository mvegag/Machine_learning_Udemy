```python
import pandas as pd
df = pd.read_csv("datos/tiendas_procesado.csv", index_col="fecha")
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use(
    "ggplot"
)

ax = df.plot.line(
    figsize=(10, 5),
    title="Venta de 3 tipos de productos en el a√±o 2015"
)

ax.set_xlabel(
    "Periodo"
)

ax.set_ylabel(
    "Ventas (en millones de d√≥lares)"
)
```

```python
import numpy as np
# hacer visible todos los ticks
# tenemos tantos ticks como filas en el dataset
plt.style.use(
    "ggplot"
)

ax = df.plot.line(
    figsize=(10, 5),
    title="Venta de 3 tipos de productos en el a√±o 2015"
)

posiciones_ticks = np.arange(0, len(df))
ax.set_xticks(
    posiciones_ticks
)

ax.set_xticklabels(
    df.index.values
)

ax.set_xlabel(
    "Periodo"
)

ax.set_ylabel(
    "Ventas (en millones de d√≥lares)"
)
```

```python
# plotly
import plotly.express as px
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template="ggplot2",
    title="Ventas de 3 tipos de productos para el 2015",
    labels={
        "fecha": "Periodo", 
        "value": "Ventas (en millones de d√≥lares)",
        "variable": "Tipo de producto"
    }
)

fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values,
    tickmode="array"
)

fig.show()
```
