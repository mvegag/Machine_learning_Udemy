```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datos/tienda_ventas.csv")
df
```

```python
df["store_nbr"].unique()
```

```python
len(df["store_nbr"].unique())
```

```python
# fechas unicas
len(df["date"].unique())
```

```python
df["mes"] = df["date"].apply(lambda x: x[:7])
df
```

```python
agrupado_tiendas = df.groupby("store_nbr").agg({
    "sales": "sum"
})
agrupado_tiendas
```

```python
agrupado_tiendas.describe()
```

```python
agrupado_tiendas.plot.box()
```

```python
agrupado_tiendas.plot.hist(bins=20)
```

```python
# enfocandose en las tiendas que mas han vendido
agrupado_tiendas.sort_values(by="sales", ascending=False)
```

```python
top_5 = [44, 45, 47, 3, 46]
df_top_5 = df[df["store_nbr"].isin(top_5)]
df_top_5
```

```python
top_5_por_mes = df_top_5.groupby(["store_nbr", "mes"]).agg({
    "sales": "sum"
})
top_5_por_mes = top_5_por_mes.reset_index()
top_5_por_mes
```

```python
sns.lineplot(
    data = top_5_por_mes,
    x = "mes", 
    y = "sales",
    hue = "store_nbr"
)
```

```python
ax = sns.lineplot(
    data = top_5_por_mes,
    x = "mes", 
    y = "sales",
    hue = "store_nbr",
    style = "store_nbr",
    markers = True,
    #palette = "Set1"
)

ax.tick_params(
    axis="x",
    labelrotation=90,
    labelsize=8
)

ax.tick_params(
    axis="y",
    labelsize=8
)

ax.grid(
    visible=True,
    axis="x",
    alpha=0.2
)

ax.grid(
    visible=True,
    axis="y",
    alpha=0.2
)

ax.legend(
    title="Top 5 tiendas",
    fontsize=8
)

ax.set_ylabel("Ventas")
ax.set_xlabel("Mes")

fig = ax.get_figure()
fig.set_size_inches(11, 3)
```

```python
top_5
```

```python
# graficacion iterativa
# color, opacidad, marcador, tipo de linea, tamaño
estilos_por_tienda = {
    44: ["grey", 0.4, "+", "--", 1],
    45: ["grey", 0.4, "s", "--", 1],
    47: ["grey", 0.4, "x", "-", 1],
    3:  ["tab:red", 1, "o", "-", 1.5],
    46: ["grey", 0.4, "D", "--", 1],
}
```

```python
fig2, ax2 = plt.subplots()

for tienda in top_5:
    estilos = estilos_por_tienda[tienda]
    df_tmp = top_5_por_mes[top_5_por_mes["store_nbr"] == tienda]

    ax2.plot(
        df_tmp["mes"],
        df_tmp["sales"],
    	label = tienda,
        markersize = 4,
        color = estilos[0],
        alpha= estilos[1],
        marker = estilos[2],
        linestyle = estilos[3],
        linewidth = estilos[4]
    )

ax2.tick_params(
    axis="x",
    labelrotation=90,
    labelsize=8
)

ax2.tick_params(
    axis="y",
    labelsize=8
)

ax2.grid(
    visible=True,
    axis="x",
    alpha=0.2
)

ax2.grid(
    visible=True,
    axis="y",
    alpha=0.2
)

ax2.legend(
    title="Top 5 tiendas",
    fontsize=8
)

ax2.set_ylabel("Ventas")
ax2.set_xlabel("Mes")

fig2.set_size_inches(11, 3)
```
