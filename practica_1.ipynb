```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datos/comparasion_algoritmos.csv")
df
```

```python
df.plot.bar(
    x="dataset",
    y=["algoritmo1", "algoritmo2"]
)
```

```python
df.plot.bar(
    x="dataset",
    y=["algoritmo1", "algoritmo2"],
    figsize=(12, 5)
)
```

```python
ax = df.plot.bar(
        x="dataset",
        y=["algoritmo1", "algoritmo2"],
        figsize=(12, 5)
)
ax.set_xlabel("Dataset")
ax.set_ylabel("Rendimiento")
ax.set_title("Evaluacin de algoritmo 1 vs algoritmo 2")
```

```python
# leyenda
ax = df.plot.bar(
        x="dataset",
        y=["algoritmo1", "algoritmo2"],
        figsize=(12, 5)
)
ax.set_xlabel("Dataset")
ax.set_ylabel("Rendimiento")
ax.set_title("Evaluacin de algoritmo 1 vs algoritmo 2")
ax.legend(
    labels=["Algoritmo 1", "Algoritmo 2"]
)
```

```python
# cuadricula
ax = df.plot.bar(
        x="dataset",
        y=["algoritmo1", "algoritmo2"],
        figsize=(12, 5)
)
ax.set_xlabel("Dataset")
ax.set_ylabel("Rendimiento")
ax.set_title("Evaluacin de algoritmo 1 vs algoritmo 2")
ax.legend(
    labels=["Algoritmo 1", "Algoritmo 2"]
)
ax.grid(visible=True)
```

```python
# cuadricula por debajo de las barras
ax = df.plot.bar(
        x="dataset",
        y=["algoritmo1", "algoritmo2"],
        figsize=(12, 5)
)
ax.set_xlabel("Dataset")
ax.set_ylabel("Rendimiento")
ax.set_title("Evaluacin de algoritmo 1 vs algoritmo 2")
ax.legend(
    labels=["Algoritmo 1", "Algoritmo 2"]
)
ax.grid(visible=True)
ax.set_axisbelow(True)
```
