```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datos/datasets_top_4.csv", index_col="dataset")
df
```

```python
# dividir en dos 
len(df)
```

```python
df1 = df.iloc[0:13, :]
df2 = df.iloc[13:, :]
len(df1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

df1.plot.barh(
    stacked=True,
    ax=ax1
)

df2.plot.barh(
    stacked=True,
    ax=ax2
)
```

```python
# reducir ancho de barras
fig, (ax1, ax2) = plt.subplots(1, 2)

df1.plot.barh(
    stacked=True,
    ax=ax1,
    width=0.3
)

df2.plot.barh(
    stacked=True,
    ax=ax2,
    width=0.3
)
```

```python
# nombre de ejes
fig, (ax1, ax2) = plt.subplots(1, 2)

df1.plot.barh(
    stacked=True,
    ax=ax1,
    width=0.3
)

df2.plot.barh(
    stacked=True,
    ax=ax2,
    width=0.3
)

ax2.set_ylabel("")
ax1.set_ylabel("Dataset")

ax2.set_xlabel("Porcentaje")
ax1.set_xlabel("Porcentaje")

ax1.legend().remove()

ax2.legend(
    ["Categoria 1", "Categoria 2", "Categoria 3", "Otros"],
    ncol=4,
    loc="upper left",
    bbox_to_anchor=(-1.3, -0.15)
)


fig.suptitle("Distribucin categoras dentro de los datasets")

# mover eje x a la derecha
ax2.yaxis.tick_right()
```

```python
# modificar colores
fig, (ax1, ax2) = plt.subplots(1, 2)

df1.plot.barh(
    stacked=True,
    ax=ax1,
    width=0.3,
    color={
        "cat1": "#000086",
        "cat2": "#5396f4",
        "cat3": "#a1dbe8",
        "otros": "#DADADA"
    }
)

df2.plot.barh(
    stacked=True,
    ax=ax2,
    width=0.3,
    color={
        "cat1": "#000086",
        "cat2": "#5396f4",
        "cat3": "#a1dbe8",
        "otros": "#DADADA"
    }
)

ax2.set_ylabel("")
ax1.set_ylabel("Dataset")

ax2.set_xlabel("Porcentaje")
ax1.set_xlabel("Porcentaje")

ax1.legend().remove()

ax2.legend(
    ["Categoria 1", "Categoria 2", "Categoria 3", "Otros"],
    ncol=4,
    loc="upper left",
    bbox_to_anchor=(-1.3, -0.15)
)


fig.suptitle("Distribucin categoras dentro de los datasets")

# mover eje x a la derecha
ax2.yaxis.tick_right()
```
