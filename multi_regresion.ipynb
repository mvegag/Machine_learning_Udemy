```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("datos/pinguinos.csv")
df = df.dropna()
df
```

```python
# grafico de regresion
sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="culmen_length_mm",
    scatter_kws={
        "s": 10
    }
)
```

```python
# coeficiente de regresion
import scipy as sp
r, p = sp.stats.pearsonr(df["flipper_length_mm"], df["culmen_length_mm"])
print(f"correlacion {r}")
```

```python
sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="culmen_length_mm",
    scatter_kws={
        "s": 10
    }
)
```

```python
sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="culmen_length_mm",
    scatter_kws={
        "s": 10
    }
)

plt.text(
    220,
    35,
    f"r = {round(r, 3)}"
)
```
