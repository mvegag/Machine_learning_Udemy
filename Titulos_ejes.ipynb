```python
# seaborn

import seaborn as sns
```

```python
plt.figure(
    figsize=(8, 4)
)

sns.set_style("ticks")
sns.set_theme("paper")

ax = sns.lineplot(
    data=df
)
```

```python
plt.figure(
    figsize=(8, 4)
)

sns.set_style("ticks")
sns.set_theme("paper")

ax = sns.lineplot(
    data=df
)

ax.set_xlabel("Periodo")
```

```python
plt.figure(
    figsize=(8, 4)
)

plt.title("Ventas de 3 tipos de productos en el año 2015") # método global
plt.xlabel("Periodo")
plt.ylabel("Ventas (en millones de dólares)")

sns.set_style("ticks")
sns.set_theme("paper")

ax = sns.lineplot(
    data=df
)
```

```python
plt.figure(
    figsize=(8, 4)
)

ax = sns.lineplot(
    data=df
)

ax.set(
    xlabel="Periodo",
    ylabel="Ventas (en millones de dólares)",
    title="Ventas de 3 tipos de productos en el año 2015"
)
```

```python
# plotly

import plotly.express as px
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template="ggplot2",
    title="Ventas de 3 tipos de productos para el 2015",
    labels={
        "fecha": "Periodo", 
        "value": "Ventas (en millones de dólares)",
        "variable": "Tipo de producto"
    }
)

fig.show()
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template="ggplot2",
    title="Ventas de 3 tipos de productos para el 2015"
)

fig.update_layout(
    xaxis_title="Periodo",
    yaxis_title="Ventas (en millones de dólares)"
)

fig.show()
```
