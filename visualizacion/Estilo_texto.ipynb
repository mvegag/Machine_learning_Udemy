```python
import pandas as pd
df = pd.read_csv("datos/tiendas_procesado.csv", index_col="fecha")
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use(
    "ggplot"
)

ax = df.plot.line(
    figsize=(10, 5),
    title="Venta de 3 tipos de productos en el a√±o 2015"
)

ax.set_xlabel(
    "Periodo"
)

ax.set_ylabel(
    "Ventas (en millones de d√≥lares)"
)
```

```python
# plotly
import plotly.express as px
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template="ggplot2",
    title="Ventas de 3 tipos de productos para el 2015"
)

fig.update_layout(
    xaxis_title="Periodo",
    yaxis_title="Ventas en millones de USD",
    font_color="blue",
    font_family="Courier new"
)

fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values,
    tickmode="array"
)

fig.update_yaxes(
    dtick=250000
)

fig.show()
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template="ggplot2",
    title="Ventas de 3 tipos de productos para el 2015"
)

fig.update_layout(
    xaxis_title="Periodo",
    yaxis_title="Ventas en millones de USD", # global
    font_color="blue",
    font_family="Courier new",
    legend_font_color="green",
    legend_title_font_color="red",
    title_font_color="violet",
    title_font_family="Times new roman"
)

fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values, # para cada eje
    tickmode="array",
    title_font_family="Arial",
    title_font_color="black",
    title_font_size=26,
    tickfont={
        "family": "Rockwell",
        "color": "brown",
        "size": 10
    }
)

fig.update_yaxes(
    dtick=250000
)

fig.show()
```
