```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("datos/paises_vida.csv")
df = df.dropna()
df = df[df["year"] == 2007]
df
```

```python
fig = px.treemap(
    df,
    path=["continent", "country"],
    values="pop",
    width=1000,
    height=800
)

fig.show()
```

```python
fig = px.treemap(
    df,
    path=["continent", "country"],
    values="pop",
    width=1000,
    height=800,
    color="lifeExp",
    color_continuous_scale="RdBu"
)

fig.show()
```

```python
fig = px.treemap(
    df,
    path=[px.Constant("Mundo"), "continent", "country"],
    values="pop",
    width=1000,
    height=800,
    color="lifeExp",
    color_continuous_scale="RdBu"
)

fig.show()
```
