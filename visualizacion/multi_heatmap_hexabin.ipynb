## heatmap categorico

```python
df = pd.read_csv("datos/tienda_ventas.csv")
df = df.dropna()
df
```

```python
df["Periodo"] = df["date"].apply(lambda x: x[0:7])
df
```

```python
df2 = df[df["store_nbr"].isin([4, 54, 55, 56, 57, 58, 59])]
df2
```

```python
df_agrupado = df[df["store_nbr"].isin([3, 44, 45, 46, 47, 48, 49, 50])].groupby(["Periodo", "store_nbr"]).agg({
    "sales": "sum"
})
df_agrupado
```

```python
df_agrupado.unstack().droplevel(0, axis=1)
```

```python
df_final = df_agrupado.unstack().droplevel(0, axis=1).transpose()
df_final
```

```python
sns.heatmap(
    data=df_final,
    cmap="Blues"
)
```

```python
sns.heatmap(
    data=df_final,
    cmap="Blues",
    square=True
)
```

```python
plt.figure(
    figsize=(15, 8)
)
sns.heatmap(
    data=df_final,
    cmap="Blues",
    square=True
)
```

```python
plt.figure(
    figsize=(15, 8)
)
sns.heatmap(
    data=df_final,
    cmap="Blues",
    square=True,
    linewidth=0.1
)
```

```python
plt.figure(
    figsize=(15, 8)
)
sns.heatmap(
    data=df_final,
    cmap="Blues",
    square=True,
    linewidth=0.1,
    annot=True
)
```

## Heatmap correlaciones

```python
df = pd.read_csv("datos/pinguinos.csv")
df = df.dropna()
df
```

```python
matriz = df[["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]].corr()
matriz
```

```python
sns.heatmap(
    matriz, 
    cmap="Blues",
    square=True,
    annot=True
)
```

```python
# paletas divergentes
sns.heatmap(
    matriz, 
    cmap="vlag",
    square=True,
    annot=True
)
```

```python
sns.heatmap(
    matriz, 
    cmap="vlag_r",
    square=True,
    annot=True
)
```

```python
sns.heatmap(
    matriz, 
    cmap="coolwarm_r",
    square=True,
    annot=True
)
```
