```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datos/titanic.csv")
df
```

```python
df = df.dropna(subset=["Age"], axis="index")
df
```

## Pandas

```python
df.boxplot(
    column="Age", 
    by="Sex"
)
```

```python
df.boxplot(
    column="Age"
)
```

## Seaborn

```python
sns.boxplot(
    data=df,
    x="Sex",
    y="Age"
)
```

# Violines

```python
sns.violinplot(
    data=df,
    x="Sex",
    y="Age"
)
```

```python
sns.violinplot(
    data=df,
    x="Sex",
    y="Age",
    inner="quartiles"
)
```

```python
sns.violinplot(
    data=df,
    x="Sex",
    y="Age",
    inner="stick"
)
```

```python
sns.violinplot(
    data=df,
    x="Sex",
    y="Age",
    inner="point"
)
```
