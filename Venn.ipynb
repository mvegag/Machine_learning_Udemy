```python
from matplotlib_venn import venn2, venn2_unweighted
from matplotlib_venn import venn3, venn3_unweighted
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("datos/animales_caracteristicas.csv")
df
```

```python
respira = set(df[df["respira"] == 1]["animal"])
respira
```

```python
piernas = set(df[df["piernas"] == 1]["animal"])
piernas
```

```python
nada = set(df[df["nada"] == 1]["animal"])
nada
```

```python
venn3(
    [
        respira,
        piernas,
        nada
    ],
    set_labels=("Respira", "Tiene piernas", "Puede nadar")
)
plt.title("Animales")
```

```python
venn3_unweighted(
    [
        respira,
        piernas,
        nada
    ],
    set_labels=("Respira", "Tiene piernas", "Puede nadar")
)
plt.title("Animales")
```

```python
venn2(
    subsets=(30, 10, 5)
)
```

```python
venn2_unweighted(
    subsets=(30, 10, 5)
)
```
