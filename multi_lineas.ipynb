```python
import pandas as pd
```

```python
df = pd.read_csv("datos/paises_vida.csv", index_col="index")
df = df.dropna()
df_asia = df[df["continent"] == "Asia"]
```

```python
agrupado = df_asia.groupby("year").agg({
    "lifeExp": "mean",
    "gdpPercap": "mean"
})
agrupado
```

```python
agrupado.plot.line(
    y=["lifeExp", "gdpPercap"],
    grid=True
)
```

```python
# lineas dobles
import matplotlib.pyplot as plt
```

```python
fig, ax1 = plt.subplots()

ax1.plot(
    agrupado.index.values,
    agrupado["gdpPercap"],
    color="tab:red",
    marker="s"
)

ax2 = ax1.twinx()

ax2.plot(
    agrupado.index.values,
    agrupado["lifeExp"],
    color="tab:blue",
    marker="o"
)

ax1.tick_params(axis="y", labelcolor="tab:red")
ax2.tick_params(axis="y", labelcolor="tab:blue")

ax1.set_ylabel("GDP per capita")
ax2.set_ylabel("Expectativa de vida")
ax1.set_xlabel("Año")
ax1.grid(True)
ax2.grid(True)

ax1.legend(
    title="",
    labels=["GDP per capita"],
    labelcolor=["blue"]
)

ax2.legend(
    bbox_to_anchor=(0.4, 0.93),
    title="",
    labels=["Expectativa de vida"],
    labelcolor=["blue"]
)
```
